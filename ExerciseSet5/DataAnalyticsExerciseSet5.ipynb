```python
url = 'https://en.wikipedia.org/wiki/Rovaniemi'
page = requests.get(url)
soup = BeautifulSoup(page.content,'html.parser')

# Get the coordinates from the Wikipedia page
coordinates = soup.find(id='coordinates').text
coordinates.replace('\ufeff','').strip()
```

```python
# Get the nickname of the city
nickname = soup.find(class_='ib-settlement-nickname nickname').text
nickname.strip()
```

```python
# Get the total population of the city
elem1 = soup.find_all('tr', class_='mergedrow')

for elem in elem1:
    if 'Total' in elem.text:
        population = elem.find('td', class_='infobox-data').text
        population = population.replace(',', '')
        if population.isnumeric():    
            print(f'Total population: {population}')
```

```python
# Get the total area of the city
for elem in elem1:
    if 'Total' in elem.text:
        area = elem.find('td', class_='infobox-data').text
        if 'km2' in area: 
            area = area.split(' ')[0]
            area = area.replace(',', '').replace('km2','')
            area = float(area)   
            print(f'Total area: {area} km2')
```

<b>Extra task: </b> Do the same but with <b>temperatures_unfixed.csv</b> instead. Remember <b>groupby()</b> before attempting merging! 

<b>Also, you might want to remove some of the columns before grouping data => <span style="color: red;">Summary + Hour</span>. </b>

<span><i>You can also try to merge the DataFrames so that the Summary –of the weather is also added. For this, you probably need to figure out first how to determine the "average" Summary for each day (extra challenge!).</i></span>


```python
# "average" summary for each day
def average_summary_all(summaries):
    avg = ''

    for summary in summaries:
        if 'Cloudy' in summary and 'Cloudy' not in avg:
            avg += 'Cloudy, '
        if 'Overcast' in summary and 'Overcast' not in avg:
            avg += 'Overcast, '
        if 'Clear' in summary and 'Clear' not in avg:
            avg += 'Clear, '
        if 'Rain' in summary and 'Rainy' not in avg:
            avg += 'Rainy, '
        if 'Drizzle' in summary and 'Rainy' not in avg:
            avg += 'Drizzle, '
        if 'Windy' in summary and 'Windy' not in avg:
            avg += 'Windy, '
        if 'Breezy' in summary and 'Windy' not in avg:
            avg += 'Breezy, '
        if 'Foggy' in summary and 'Foggy' not in avg:
            avg += 'Foggy, '
        if 'Humid' in summary and 'Humid' not in avg:
            avg += 'Humid, '
        if 'Dry' in summary and 'Dry' not in avg:
            avg += 'Dry, '
    
    return avg[:-2]
        
```

```python
df_temperatures_unfixed = pd.read_csv('temperatures_unfixed.csv')

# Clean up the temperature data
df_temperatures_unfixed.drop(columns=['Hour'], inplace=True)

# Group by date and calculate the mean temperature
df_temperatures_unfixed = df_temperatures_unfixed.groupby('Date').agg({'Temperature (C)': 'mean', 'Summary': average_summary_all}).reset_index()

# Merge the dataframes
df_merged2 = df_coffee_sales.merge(df_temperatures_unfixed, left_on='Date', right_on='Date', how='left')
df_merged2.head()
```
